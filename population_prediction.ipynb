```python
import regression as LR
import pandas as pd
```

```python
population_India_dataset = pd.read_csv('IndiaPopulation.csv')
```

## Lets's predict the population from 2020 to 2030 using the linear regression algorithm created from scratch

```python
years = list(range(2020,2030))
```

```python
predModel = LR.predict(population_India_dataset, years, 'population_india_dataset')
predModel.to_csv('./files/population_india_dataset.csv', index = False)
```
